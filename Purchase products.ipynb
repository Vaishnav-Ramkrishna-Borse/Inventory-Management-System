```python
ui_name = input("Enter Your Name: ")
p = input("Please enter Your phone no.")
if len(p) in range(10,12):
    ph_no = int(p)
else :
    ph_no = "Skipped"
    
fd = open("record.json",'r')
r = fd.read()
fd.close()

import json
record = json.loads(r)
print(record)

ui_prod = input("Enter the product id: ")
ui_quant = int(input("Enter the quantity: "))

if ui_prod in record.keys():
    if record[ui_prod]["qn"] <= 0:
        print(record[ui_prod]["name"] +' '+ "is out of stock")
    elif ui_quant <= record[ui_prod]["qn"] and ui_quant > 0:
        print("*******************************************")
        print("User Name: ",ui_name)
        print("User ph no. :",ph_no)
        print("*******************************************")
        print("Product: ",record[ui_prod]["name"])
        print("Weight per quantity :",record[ui_prod]["wt"])
        print("mfg.date: ",record[ui_prod]["mfd"])
        print("exp.date: ",record[ui_prod]["exd"])
        print("Price: ",record[ui_prod]["pr"])
        print("Quantity: ",ui_quant)
        print("-------------------------------------------")
        print("Billing Amount: ",record[ui_prod]["pr"] * ui_quant) 
        print("*******************************************")
        
        record[ui_prod]["qn"] = record[ui_prod]["qn"] - ui_quant
        
        import time
        t = time.ctime()

        js = json.dumps(record)

        fd = open("record.json",'w')
        fd.write(js)
        fd.close()

        fd = open("sales.json", "r")
        SALE = fd.read()
        fd.close()

        if len(SALE) == 0:
            SALE = "{}"
        else:
            pass

        sales = json.loads(SALE)

        if len(sales) == 0:
            sales = {1 : {'prod': ui_prod, 'qn': ui_quant, 'amount': record[ui_prod]['pr'] * ui_quant, 'time': t,"name": ui_name , 'phone': ph_no}}
        else:
            sales[len(sales)+1] = {'prod': ui_prod, 'qn': ui_quant, 'amount': record[ui_prod]['pr'] * ui_quant, 'time': t, "name": ui_name, 'phone': ph_no}
    
        sale = json.dumps(sales)
        fd = open("sales.json","w")
        fd.write(sale)
        fd.close()
    else:
        print("Enter the quantity less than",record[ui_prod]["qn"] + 1)
else:
    print("Plz Enter valid product id")
```
